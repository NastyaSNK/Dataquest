```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads=pd.read_csv("recent-grads.csv")
recent_grads.iloc[1]
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
raw_data_count=recent_grads.shape[0]
print(raw_data_count)
recent_grads=recent_grads.dropna(axis=0,inplace=False)
raw_data_count=recent_grads.shape[0]
print(raw_data_count)
```

# Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Employed vs. Sample_size')
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Employed vs. Sample_size')
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Employed vs. Sample_size')
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Employed vs. Sample_size')
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Employed vs. Sample_size')
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Employed vs. Sample_size')
```

# Pandas, Histograms¶

```python
cols=["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]
fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=40)
    ax.set_xlabel(cols[r])
    
```

```python
fig = plt.figure(figsize=(5,25))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=45)
    ax.set_xlabel(cols[r])
```
